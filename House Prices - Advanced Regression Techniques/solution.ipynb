```python
def get_dataset():
    import os
    import kaggle as kg
    import shutil

    try:
        from google.colab import drive, userdata
        drive.mount('/content/drive')

        IN_COLAB = True
        DATA_PATH = '/content/drive/My Drive/Datasets/House Prices - Advanced Regression Techniques'
        os.environ['KAGGLE_API_TOKEN'] = userdata.get('KAGGLE_API_TOKEN')
    except:
        import dotenv
        IN_COLAB = False
        DATA_PATH = './data'
        dotenv.load_dotenv('../.env')

    if os.path.exists(DATA_PATH):
        shutil.rmtree(DATA_PATH)

    kg.api.authenticate()

    if not os.path.exists(DATA_PATH):
        os.mkdir(DATA_PATH)

    import zipfile
    kg.api.competition_download_files(competition='house-prices-advanced-regression-techniques', path=DATA_PATH, quiet=False)
    for file in os.listdir(DATA_PATH):
        print(f"Extracting files for {file}")
        with zipfile.ZipFile(f'{DATA_PATH}/{file}', 'r') as z:
            z.extractall(DATA_PATH)
        os.remove(f'{DATA_PATH}/{file}')
        print(f"Extraction done for {file}, deleting...")

```
